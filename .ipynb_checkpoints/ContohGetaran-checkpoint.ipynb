```python
import pandas as pd
```

```python
y = []
for x in ['tanahB2B.csv','tanahB2K.csv','tanahK2K.csv']:
    y.append(pd.read_csv(x))
df = pd.concat(y)
```

```python
print(df.columns)
```

```python
data = df[['humidity1', 'vibration1', 'humidity2',
       'vibration2', 'temperature', 'distance']].to_numpy()
```

```python
label = df['status'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(data, label, test_size=0.10, random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC(gamma='scale')
clf.fit(train_x,train_y)
```

```python
pred_y = clf.predict(test_x)
from sklearn.metrics import accuracy_score
score = accuracy_score(test_y,pred_y)
print(score)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import svm
X = df[['vibration1','vibration2']].to_numpy()
y = df['status'].map({'basah ke basah': 0,
                      'basah ke kering': 1,'kering ke kering':2})
    
clf = svm.SVC(kernel="sigmoid", C=1000)
clf.fit(X, y)
plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)

ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn import datasets, svm
from sklearn.inspection import DecisionBoundaryDisplay

# import some data to play with
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000, dual="auto"),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```
